### Reorganizing Arrays

#### .reshape()
`reshape()`

````python
before = np.array([[1,2,3,4],[5,6,7,8]])
print("before")
print(before)
print("shape:", before.shape)

print('after:')
after = before.reshape((4,2))
print(after)
print("shape:", after.shape)

print('after:')
after = before.reshape((2,2,2))
print(after)
print("shape:", after.shape)
````

````python
print('after:')
after = before.reshape((1,-1))
print(after)
print("shape:", after.shape)
````

````python
def board(n):
    print ("board 8x8")
    x = np.zeros((n, n), dtype = int)
    x[1::2, ::2] = 1
    x[::2, 1::2] = 1
    for i in range (n):
        for j in range (n):
            print (x[i][j], end = " ")
        print()
n = 8
board(n)
````
