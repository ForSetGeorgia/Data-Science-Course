```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.__version__
```

## Load Data

### Simpsons

```python
simpsons_df = pd.read_csv('data_simpsons_episodes.csv')
simpsons_df.head()
```

### Drinks

```python
drinks_df = pd.read_excel('drinks.xlsx')
drinks_df.head()
```

### Titanic

```python
titanic_df = pd.read_csv('data-titanic.csv')
```

## Plots

```python
sns.distplot(drinks_df['beer_servings'])
```

### displot
[.displot()](https://seaborn.pydata.org/generated/seaborn.displot.html)

```python
sns.displot(data=drinks_df, x='beer_servings')
```

```python
sns.displot(data=drinks_df, x='beer_servings', kde=True)
```

```python
sns.displot(data=drinks_df, x= 'beer_servings', kind="kde")
```

```python
sns.displot(data=drinks_df, x= 'beer_servings', kind="kde", rug=True)
```

### lmplot
[lmplot](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

```python
sns.lmplot(x='beer_servings' , y='wine_servings', data=drinks_df)
```

### jointplot
[.jointplot()](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

```python
sns.jointplot(data=drinks_df, x="beer_servings", y="wine_servings")
```

### boxplot
[.boxplot()](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
sns.boxplot(data=drinks_df)
```

### violinplot
[.violinplot()](https://seaborn.pydata.org/generated/seaborn.violinplot.html)

```python
simpsons_season_24 = simpsons_df.loc[simpsons_df['season']==24,['us_viewers_in_millions']]
simpsons_season_24.head()
```

```python
height = 8
width = 20

sns.displot(data=simpsons_season_24, x='us_viewers_in_millions', height=height, aspect=width/height, kind="kde")
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=100)

sns.violinplot(x='season', y='us_viewers_in_millions', data=simpsons_df)

plt.show()
```

### swarmplot
[.swarmplot()](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)

```python
plt.figure(figsize=(10,4), dpi=100)

sns.swarmplot(x='season', y='us_viewers_in_millions', data=simpsons_df, size=2)

plt.show()
```

### stripplot
[.stripplot()](https://seaborn.pydata.org/generated/seaborn.stripplot.html)

```python
plt.figure(figsize=(10,4), dpi=100)

sns.stripplot(x='season', y='us_viewers_in_millions', data=simpsons_df)

plt.show()
```

### barplot
[.barplot()](https://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
plt.figure(figsize=(10,4), dpi=100)

sns.barplot(x='season', y='us_viewers_in_millions', data=simpsons_df)

plt.show()
```

### FacetGrid
[.FacetGrid()](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html)

```python
g = sns.FacetGrid(titanic_df, col='Sex', hue='Survived')
g.map(plt.hist, "Age", alpha=0.5)
g.add_legend()
```

```python
g = sns.FacetGrid(titanic_df, row='Sex', hue='Survived')
g.map(plt.hist, "Age", alpha=0.5)
g.add_legend()
```

```python
g = sns.FacetGrid(titanic_df, col='Sex', row='Embarked', hue='Survived')
g.map(plt.hist, "Age", alpha=0.5)
g.add_legend()
```

```python
g = sns.FacetGrid(titanic_df, row='Sex', col='Embarked', hue='Survived')
g.map(plt.hist, "Age", alpha=0.5)
g.add_legend()
```
